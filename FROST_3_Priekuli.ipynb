```python
# Izveido Mekaiņu sensoru apvionoto failu

path = '/content/'

name_list = []
full_list = []

for f in os.listdir(path):
  if f.endswith('.csv'):
     name_list.append(f)
# print(name_list)

for f in name_list:
    h = path + f
    full_list.append(h)
# # print(full_list)

# vides_liste = ["Mežs", "Pļava", "Mežs_M", "Mežs_M", "Ceļš", "Pļava", "Ceļš"]

# vides_liste = ["Sauss lauks kalna gals", "Sauss dumbrājs Mežā", "Sauss mežs, apkārt slapjš kūdrenis", "Slapja vieta dumbrāja/Meža malā, kur pāreja uz sausu", "kūdrenis, daļēji mitra vieta ieplakā", "Sauss māls kalna galā/atverta vieta [ceļš?]", "Mitrs lauks, ieplaka", "Sauss mežs, kalna galā"]

vides_liste = ["Sauss_lauks", "Mežs", "Mežs", "Mežs_M", "Slapja_atklāta_vieta", "Sausa_atklāta_vieta", "Mitrs_lauks", "Mežs"]

comb = pd.DataFrame()

for path, vide in zip(full_list, vides_liste):
  dfg = pd.read_csv(path)
  dfg['date'] = pd.to_datetime(dfg.datetime)

  dfg = dfg[(dfg['date']> "2021-01-01")]
  # dfg = dfg.loc[dfg['coordinate_id'] == list(dfg['coordinate_id'].unique())[0]]
  dfg["Vide"] = vide
  dfg['name'] = dfg["Vide"]+str('_')+str(path[9:13])

  comb = pd.concat([comb, dfg])

comb

comb.to_csv(r'/content/OneSoil_Mekaini_sensors.csv')
```
